**Praktikum 01: Enkapsulasi pada kelas Bank Account**

```python
class BankAccount:
    def __init__(self, owner, balance):
        # Atribut private (menggunakan double underscore)
        self.__owner = owner
        self.__balance = balance

    def deposit(self, amount):
        """Method untuk menambahkan saldo."""
        if amount > 0:
            self.__balance += amount
            print(f"{amount} telah ditambahkan ke akun {self.__owner}.")
        else:
            print("Jumlah deposit harus lebih dari 0.")

    def withdraw(self, amount):
        """Method untuk menarik saldo."""
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} telah ditarik dari akun {self.__owner}.")
        else:
            print("Saldo tidak mencukupi.")

    def get_balance(self):
        """Method untuk mendapatkan saldo terkini."""
        return self.__balance


# Contoh penggunaan
if __name__ == "__main__":
    # Membuat objek BankAccount
    alice_account = BankAccount(owner="Alice", balance=1000)

    # Deposit uang
    alice_account.deposit(500)    # Berhasil
    alice_account.deposit(-100)   # Gagal

    # Withdraw uang
    alice_account.withdraw(300)   # Berhasil
    alice_account.withdraw(2000)  # Gagal

    # Mendapatkan saldo
    current_balance = alice_account.get_balance()
    print(f"Saldo terakhir di akun Alice: {current_balance}")

    # Akses atribut private (tidak direkomendasikan)
    # print(alice_account.__balance)  # Error

    # Python akan melakukan name mangling:
    # __balance menjadi _BankAccount__balance
    print(alice_account._BankAccount__balance)  # Bisa diakses, tapi tidak disarankan
```

**Praktikum 02: Enkapsulasi pada kelas Employee**

```python
class Employee:
    def __init__(self, name, salary):
        # Atribut private
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase):
        """Method untuk menaikkan gaji dengan validasi."""
        if increase > 0:
            self.__salary += increase
            print(f"Gaji telah dinaikkan sebesar {increase}.")
        else:
            print("Nilai kenaikan harus lebih dari 0.")

    def set_salary(self, new_salary):
        """Method untuk mengubah gaji dengan validasi."""
        if new_salary >= 0:
            self.__salary = new_salary
            print(f"Gaji diatur ulang menjadi {new_salary}.")
        else:
            print("Gaji tidak dapat bernilai negatif.")

    def get_salary(self):
        """Method untuk mendapatkan informasi gaji."""
        return self.__salary

    def get_employee_info(self):
        """Method untuk menampilkan informasi karyawan secara menyeluruh."""
        return f"Employee: {self.__name}, Gaji: {self.__salary}"


# Contoh penggunaan
if __name__ == "__main__":
    # Membuat objek Employee
    employee1 = Employee("John Doe", 50000)

    # Tampilkan informasi awal
    print(employee1.get_employee_info())

    # Update gaji
    employee1.update_salary(5000)
    print(f"Gaji setelah kenaikan: {employee1.get_salary()}")

    # Set ulang gaji
    employee1.set_salary(60000)
    print(f"Informasi terbaru: {employee1.get_employee_info()}")

    # Akses langsung (tidak disarankan)
    # print(employee1.__salary)  # Error

    # Name mangling (jalan belakang)
    print(employee1._Employee__salary)  # Bisa, tapi jangan dipakai di real project
```

**Praktikum 03: Konstruktor dan Destruktor Sederhana**

```python
class SimpleExample:
    def __init__(self, name):
        """
        Konstruktor: Dipanggil saat objek dibuat.
        Menyimpan nilai 'name' dan mencetak pesan pembuatan objek.
        """
        self.name = name
        print(f"Konstruktor: Objek '{self.name}' telah dibuat.")

    def __del__(self):
        """
        Destruktor: Dipanggil saat objek dihapus.
        Mencetak pesan bahwa objek sedang dihapus.
        """
        print(f"Destruktor: Objek '{self.name}' sedang dihapus.")


def main():
    print("Program dimulai.\n")

    # Membuat objek
    obj = SimpleExample("Demo")
    print("Program sedang berjalan...\n")

    # Menghapus objek secara eksplisit
    del obj
    print("Objek telah dihapus secara eksplisit.\n")

    print("Program selesai.")


if __name__ == "__main__":
    main()
```

**Praktikum 04: Konstruktor dan Destruktor program FileLogger**

```python
class FileLogger:
    def __init__(self, filename):
        """
        Konstruktor: Membuka file log untuk menulis pesan.
        Parameter:
        - filename: Nama file tempat pesan log akan ditulis.
        """
        self.filename = filename
        try:
            self.file = open(filename, "a")
            print(f"File '{filename}' berhasil dibuka untuk logging.")
        except Exception as e:
            self.file = None
            print(f"Gagal membuka file '{filename}': {e}")

    def write_log(self, message):
        """
        Menulis pesan log ke dalam file.
        Parameter:
        - message: Pesan yang akan ditulis ke file.
        """
        if self.file:
            self.file.write(message + "\n")
            self.file.flush()  # langsung ditulis ke disk
            print(f"Pesan log: '{message}' telah ditulis.")
        else:
            print("File tidak tersedia untuk logging.")

    def __del__(self):
        """
        Destruktor: Menutup file log ketika objek dihapus.
        """
        if hasattr(self, "file") and self.file and not self.file.closed:
            self.file.close()
            print(f"File '{self.filename}' telah ditutup.")


# Contoh penggunaan
if __name__ == "__main__":
    # Membuat logger
    logger = FileLogger("application.log")

    # Menulis log
    logger.write_log("Aplikasi dimulai.")
    logger.write_log("Melakukan operasi A...")
    logger.write_log("Operasi A selesai.")
    logger.write_log("Aplikasi akan segera selesai.")

    # Menghapus objek secara eksplisit
    del logger

    # Kalau tidak pakai del, __del__ akan dipanggil saat program selesai
```

**Praktikum 05: Properti "self" pada kelas**

```python
class Calculator:
    def __init__(self, initial_value=0):
        """
        Konstruktor kelas Calculator.
        - self: Mengacu pada instance yang dibuat.
        - initial_value: Nilai awal dari kalkulator.
        """
        self.value = initial_value
        print(f"Kalkulator diinisialisasi dengan nilai: {self.value}")

    def add(self, number):
        """
        Menambahkan 'number' ke nilai yang tersimpan.
        """
        self.value += number
        print(f"Setelah penambahan {number}, nilai sekarang adalah: {self.value}")

    def subtract(self, number):
        """
        Mengurangi 'number' dari nilai yang tersimpan.
        """
        self.value -= number
        print(f"Setelah pengurangan {number}, nilai sekarang adalah: {self.value}")

    def reset(self):
        """
        Mengatur ulang nilai ke 0.
        """
        self.value = 0
        print("Nilai telah direset ke 0.")

    def show_value(self):
        """
        Menampilkan nilai saat ini.
        """
        print(f"Nilai saat ini adalah: {self.value}")


# Contoh penggunaan
def main():
    # Objek pertama
    calc1 = Calculator(initial_value=10)
    calc1.add(5)
    calc1.subtract(3)
    calc1.show_value()

    print()  # spasi

    # Objek kedua
    calc2 = Calculator()
    calc2.add(20)
    calc2.show_value()

    # Penjelasan:
    # calc1 dan calc2 punya value masing-masing (karena self berbeda)


if __name__ == "__main__":
    main()
```

Praktikum 06: Menggunakan getter, setter, dan dekorator @property untuk mengelola akses atribut

```python
class Person:
    def __init__(self, name, age):
        """
        Konstruktor untuk menginisialisasi objek Person.
        """
        self.__name = name
        self.__age = age

    # ===== NAME =====
    @property
    def name(self):
        """Getter untuk name"""
        return self.__name

    @name.setter
    def name(self, value):
        """Setter untuk name dengan validasi"""
        if not value:
            print("Nama tidak boleh kosong.")
        else:
            self.__name = value

    # ===== AGE =====
    @property
    def age(self):
        """Getter untuk age"""
        return self.__age

    @age.setter
    def age(self, value):
        """Setter untuk age dengan validasi"""
        if value < 0:
            print("Umur tidak boleh negatif!")
        else:
            self.__age = value


# Contoh penggunaan
def main():
    p1 = Person("Budi", 20)

    print(p1.name)  # Getter
    print(p1.age)

    # Setter (valid)
    p1.name = "Andi"
    p1.age = 25

    print(p1.name)
    print(p1.age)

    # Setter (invalid)
    p1.name = ""
    p1.age = -5


if __name__ == "__main__":
    main()
```

# **PENUGASAN**

**Buatlah sebuah kelas Python bernama Student.**

```python
class Student:
    def __init__(self, name, score):
        # atribut privat
        self.__name = name
        self.__score = 0
        self.__grade = ""

        # pakai setter supaya validasi & update grade jalan
        self.score = score

    # ===== NAME =====
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if not value:
            print("Nama tidak boleh kosong.")
        else:
            self.__name = value

    # ===== SCORE =====
    @property
    def score(self):
        return self.__score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()  # otomatis update grade
        else:
            print("Score harus antara 0 sampai 100.")

    # ===== METHOD INTERNAL =====
    def __update_grade(self):
        if 90 <= self.__score <= 100:
            self.__grade = "A"
        elif 80 <= self.__score <= 89:
            self.__grade = "B"
        elif 70 <= self.__score <= 79:
            self.__grade = "C"
        elif 60 <= self.__score <= 69:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # ===== METHOD TAMPIL =====
    def show_info(self):
        print(f"Nama Mahasiswa: {self.__name}")
        print(f"Nilai: {self.__score}")
        print(f"Grade: {self.__grade}")

    # ===== DESTRUKTOR =====
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")


# ===== MAIN PROGRAM =====
def main():
    # buat objek
    s = Student("Siti", 87)
    s.show_info()

    print("\nNilai diubah...\n")

    # ubah nilai → grade otomatis berubah
    s.score = 93
    s.show_info()

    # hapus objek
    del s


if __name__ == "__main__":
    main()
```
