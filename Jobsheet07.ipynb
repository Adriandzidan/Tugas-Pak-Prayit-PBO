```python
class DataStore:
    def __init__(self, nama_store="Default Store"):
        # Dictionary internal untuk menyimpan data
        self._data = {}
        self.nama = nama_store
        print(f"DataStore '{self.nama}' dibuat.")

    def tambah_data(self, key, value):
        """Menambahkan data ke dalam store."""
        print(f"[{self.nama}] Menambahkan data: '{key}' -> {value}")
        self._data[key] = value
        print(" -> Data berhasil ditambahkan.")

    def get_data(self, key):
        """
        Mengambil data berdasarkan key.
        Menangani KeyError jika key tidak ditemukan.
        """
        print(f"[{self.nama}] Mencoba mengambil data dengan key: '{key}'")

        try:
            # Baris ini bisa menimbulkan KeyError
            nilai = self._data[key]
            print(f" -> Data ditemukan: {nilai}")
            return nilai

        except KeyError:
            print(
                f" -> ERROR: Key '{key}' tidak ditemukan "
                f"di DataStore '{self.nama}'."
            )
            return None


# =========================
# Kode Utama
# =========================
if __name__ == "__main__":

    # Membuat instance DataStore
    store_produk = DataStore("Toko Produk")

    # Menambahkan beberapa data
    store_produk.tambah_data("Buku Tulis", 15000)
    store_produk.tambah_data("Pensil 2B", 2000)

    # Mengambil data yang ada
    print("\nMengambil data yang ADA:")

    harga_buku = store_produk.get_data("Buku Tulis")

    if harga_buku is not None:
        print(f"Harga Buku Tulis adalah: {harga_buku}")

    # Mengambil data yang tidak ada
    print("\nMengambil data yang TIDAK ADA:")

    harga_penggaris = store_produk.get_data("Penggaris")

    if harga_penggaris is None:
        print("Harga Penggaris tidak ditemukan (sesuai ekspektasi).")
```

```python
class KalkulatorAngka:
    def __init__(self):
        print("Kalkulator Angka siap.")

    def bagi(self, angka1, angka2):
        """
        Membagi angka1 dengan angka2.
        Menangani:
        - ValueError jika input bukan angka
        - ZeroDivisionError jika pembagi bernilai nol
        """

        print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")

        try:
            # Konversi input ke float
            num1 = float(angka1)
            num2 = float(angka2)

            # Operasi pembagian
            hasil = num1 / num2

            print(f" -> Hasil pembagian: {hasil}")
            return hasil

        except ValueError:
            print(
                " -> ERROR: Input tidak valid! "
                "Pastikan kedua input adalah angka."
            )
            return None

        except ZeroDivisionError:
            print(
                " -> ERROR: Pembagian dengan nol tidak diperbolehkan!"
            )
            return None


# =========================
# Kode Utama
# =========================
if __name__ == "__main__":

    calc = KalkulatorAngka()

    print("\n--- Tes Kalkulator ---")

    hasil1 = calc.bagi(100, 5)              # Operasi sukses
    hasil2 = calc.bagi(20, 0)               # Memicu ZeroDivisionError
    hasil3 = calc.bagi("limapuluh", 10)     # Memicu ValueError pada angka1
    hasil4 = calc.bagi(50, "sepuluh")       # Memicu ValueError pada angka2

    print("\n--- Ringkasan Hasil ---")

    print(f"Hasil 1 (100/5): {hasil1}")
    print(f"Hasil 2 (20/0): {hasil2}")
    print(f"Hasil 3 ('limapuluh'/10): {hasil3}")
    print(f"Hasil 4 (50/'sepuluh'): {hasil4}")
```

3. **Praktikum 3: Menggunakan else dan finally
dalam Metode**

```python
import os


class FileProcessor:
    def __init__(self, nama_file):
        self.nama_file = nama_file
        self._file_handle = None

        print(f"\n-> FileProcessor dibuat untuk file: '{self.nama_file}'")

    def baca_dan_proses(self):
        """
        Membuka, membaca, dan memproses file.
        Menggunakan try-except-else-finally.
        """

        print(f"[{self.nama_file}] Mencoba memproses...")

        try:
            # Membuka file
            print(f" [{self.nama_file}] Membuka file...")
            self._file_handle = open(
                self.nama_file,
                "r",
                encoding="utf-8"
            )

            # Membaca isi file
            print(f" [{self.nama_file}] Membaca isi file...")
            isi = self._file_handle.read()

            print(" -> File berhasil dibaca.")

            # Simulasi proses isi file
            print(f" [{self.nama_file}] Memproses isi...")
            panjang_isi = len(isi)

            print(f" -> Panjang isi: {panjang_isi} karakter.")

        except FileNotFoundError:
            print(
                f" -> ERROR: File '{self.nama_file}' tidak ditemukan."
            )

        except UnicodeDecodeError:
            print(
                f" -> ERROR: Gagal membaca '{self.nama_file}' "
                f"sebagai teks."
            )

        except Exception as e:
            print(
                f" -> ERROR Lainnya saat memproses file: "
                f"{type(e).__name__} - {e}"
            )

        else:
            # Dijalankan jika tidak ada error
            print(
                f" -> PEMROSESAN BERHASIL: "
                f"Tidak ada error pada '{self.nama_file}'."
            )

        finally:
            # Selalu dijalankan
            print(f" [{self.nama_file}] Blok Finally: Cleanup...")

            if self._file_handle and not self._file_handle.closed:
                print(f" Menutup file '{self.nama_file}'...")
                self._file_handle.close()

                print(" -> File berhasil ditutup.")

            else:
                print(
                    " -> File tidak perlu ditutup "
                    "(belum dibuka atau sudah ditutup)."
                )

            print(f"[{self.nama_file}] Selesai mencoba proses.")
            print("-" * 30)


# =========================
# Kode Utama
# =========================
if __name__ == "__main__":

    NAMA_FILE_SUKSES = "data_penting.txt"
    NAMA_FILE_GAGAL = "file_yang_tidak_ada.log"

    # Membuat file dummy
    try:
        with open(NAMA_FILE_SUKSES, "w", encoding="utf-8") as f:
            f.write("Ini adalah baris pertama.\n")
            f.write("Baris kedua berisi data penting.\n")

        print(f"File dummy '{NAMA_FILE_SUKSES}' berhasil dibuat.")

    except IOError as e:
        print(
            f"Gagal membuat file dummy "
            f"'{NAMA_FILE_SUKSES}': {e}"
        )

    # =========================
    # Skenario 1: File berhasil dibaca
    # =========================
    print("\n--- Skenario 1: Proses Sukses ---")

    processor1 = FileProcessor(NAMA_FILE_SUKSES)
    processor1.baca_dan_proses()

    # =========================
    # Skenario 2: File tidak ditemukan
    # =========================
    print("\n--- Skenario 2: File Tidak Ditemukan ---")

    processor2 = FileProcessor(NAMA_FILE_GAGAL)
    processor2.baca_dan_proses()

    # =========================
    # Cleanup file dummy
    # =========================
    try:
        if os.path.exists(NAMA_FILE_SUKSES):
            os.remove(NAMA_FILE_SUKSES)

            print(
                f"\nFile dummy '{NAMA_FILE_SUKSES}' berhasil dihapus."
            )

    except OSError as e:
        print(
            f"\nGagal menghapus file dummy "
            f"'{NAMA_FILE_SUKSES}': {e}"
        )
```

4. **Praktikum 4: Mengakses Objek Exception (as e)**

```python
class KalkulatorAngka:

    def __init__(self):
        print("Kalkulator Angka siap.")

    def bagi(self, angka1, angka2):
        """
        Membagi angka1 dengan angka2.
        Menangkap objek exception untuk menampilkan detail error.
        """

        print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")

        try:
            # Konversi input ke float
            num1 = float(angka1)
            num2 = float(angka2)

            # Cek pembagian dengan nol
            if num2 == 0:
                raise ZeroDivisionError(
                    "Pembagi tidak boleh nol secara eksplisit."
                )

            # Operasi pembagian
            hasil = num1 / num2

            print(f" -> Hasil pembagian: {hasil}")
            return hasil

        # Menangkap ValueError
        except ValueError as e_val:
            print(" -> ERROR Konversi: Input tidak valid.")

            # Menampilkan detail error
            print(f" Detail Error (ValueError): {e_val}")

            return None

        # Menangkap ZeroDivisionError
        except ZeroDivisionError as e_zero:
            print(
                " -> ERROR Pembagian: "
                "Tidak dapat membagi dengan nol."
            )

            # Menampilkan detail error
            print(f" Detail Error (ZeroDivisionError): {e_zero}")

            return None

        # Menangkap error lainnya
        except Exception as e:
            print(
                " -> ERROR Lainnya: "
                "Terjadi kesalahan tak terduga."
            )

            # Menampilkan tipe dan isi error
            print(
                f" Detail Error (Lainnya): "
                f"{type(e).__name__} - {e}"
            )

            return None


# =========================
# Kode Utama
# =========================
if __name__ == "__main__":

    calc = KalkulatorAngka()

    print("\n--- Tes Kalkulator ---")

    hasil1 = calc.bagi(100, 5)            # Sukses
    hasil2 = calc.bagi(20, 0)             # ZeroDivisionError
    hasil3 = calc.bagi("limapuluh", 10)   # ValueError
    hasil4 = calc.bagi(10, [])            # TypeError

    print("\n--- Ringkasan Hasil ---")

    print(f"Hasil 1 (100/5): {hasil1}")
    print(f"Hasil 2 (20/0): {hasil2}")
    print(f"Hasil 3 ('limapuluh'/10): {hasil3}")
    print(f"Hasil 4 (10/[]): {hasil4}")
```

```python
# =========================================
# Helper Format Rupiah
# =========================================
import locale

try:
    locale.setlocale(locale.LC_ALL, "id_ID.UTF-8")

except locale.Error:
    print("Locale id_ID tidak tersedia.")

    def format_rupiah(angka):
        return f"Rp {angka:,.2f}"

else:
    def format_rupiah(angka):
        return locale.currency(
            angka,
            grouping=True,
            symbol="Rp "
        )


# =========================================
# Class PengelolaAkun
# =========================================
class PengelolaAkun:

    def __init__(self, nomor_akun, saldo_awal=0):

        print(f"\nMencoba membuat akun {nomor_akun}...")

        # Validasi saldo awal
        if not isinstance(saldo_awal, (int, float)) or saldo_awal < 0:

            raise ValueError(
                f"Saldo awal ({saldo_awal}) tidak valid. "
                f"Harus angka non-negatif."
            )

        self.nomor_akun = nomor_akun
        self._saldo = float(saldo_awal)

        print(
            f" -> Akun {self.nomor_akun} berhasil dibuat "
            f"dengan saldo: {format_rupiah(self._saldo)}"
        )

    def get_saldo(self):
        """Mengembalikan saldo saat ini."""
        return self._saldo

    def tarik_tunai(self, jumlah):
        """
        Menarik tunai dengan validasi.
        Me-raise exception jika gagal.
        """

        print(
            f"\n[{self.nomor_akun}] "
            f"Mencoba tarik tunai: {format_rupiah(jumlah)}"
        )

        # Validasi jumlah penarikan
        if not isinstance(jumlah, (int, float)) or jumlah <= 0:

            raise ValueError(
                f"Jumlah penarikan ({jumlah}) tidak valid. "
                f"Harus angka positif."
            )

        # Validasi saldo cukup
        if jumlah > self._saldo:

            raise RuntimeError(
                f"Saldo tidak mencukupi! "
                f"Saldo: {format_rupiah(self._saldo)}, "
                f"Diminta: {format_rupiah(jumlah)}"
            )

        # Proses penarikan
        self._saldo -= jumlah

        print(f" -> Penarikan {format_rupiah(jumlah)} berhasil.")
        print(f" -> Saldo sekarang: {format_rupiah(self._saldo)}")

        return True


# =========================================
# Kode Utama
# =========================================
if __name__ == "__main__":

    # =====================================
    # Skenario 1: Saldo awal tidak valid
    # =====================================
    print("\n--- Skenario 1: Saldo Awal Tidak Valid ---")

    try:
        akun_gagal = PengelolaAkun("ACC001", -50000)

    except ValueError as e:
        print(f" ERROR DITANGKAP saat pembuatan akun: {e}")

    # =====================================
    # Skenario 2: Akun valid dan transaksi
    # =====================================
    print("\n--- Skenario 2: Akun Valid dan Transaksi ---")

    try:
        akun_sukses = PengelolaAkun("ACC002", 100000)

        # Transaksi sukses
        akun_sukses.tarik_tunai(30000)

        # Transaksi gagal karena negatif
        print("\n Mencoba tarik jumlah negatif...")
        akun_sukses.tarik_tunai(-10000)

    except ValueError as e:
        print(f" ERROR DITANGKAP (ValueError): {e}")

    except RuntimeError as e:
        print(f" ERROR DITANGKAP (RuntimeError): {e}")

    # =====================================
    # Skenario 3: Saldo tidak cukup
    # =====================================
    print("\n--- Skenario 3: Saldo Tidak Cukup ---")

    try:
        print(
            f" Saldo akun ACC002 saat ini: "
            f"{format_rupiah(akun_sukses.get_saldo())}"
        )

        print(" Mencoba tarik jumlah > saldo...")

        # Saldo tersisa 70000
        akun_sukses.tarik_tunai(80000)

    except ValueError as e:
        print(f" ERROR DITANGKAP (ValueError): {e}")

    except RuntimeError as e:
        print(f" ERROR DITANGKAP (RuntimeError): {e}")

    # =====================================
    # Saldo akhir
    # =====================================
    print(
        f"\nSaldo akhir akun ACC002: "
        f"{format_rupiah(akun_sukses.get_saldo())}"
    )
```

6. **Praktikum 6: Kombinasi Menangkap dan

```python
# =========================================
# Custom Exception
# =========================================
class KonfigurasiError(Exception):
    """
    Exception khusus untuk menandakan
    error terkait konfigurasi aplikasi.
    """

    def __init__(self, message, penyebab_asli=None):
        super().__init__(message)

        # Menyimpan error asli jika ada
        self.penyebab_asli = penyebab_asli

    def __str__(self):

        if self.penyebab_asli:
            return (
                f"{super().__str__()} "
                f"(Disebabkan oleh: "
                f"{type(self.penyebab_asli).__name__})"
            )

        return super().__str__()


# =========================================
# Kelas Pengelola Setting
# =========================================
class PengelolaSetting:

    def __init__(self, file_config):

        self._config = {}
        self.file_config = file_config

        # Memanggil method internal
        try:
            self._load_config()

            print(
                f"-> PengelolaSetting berhasil dibuat "
                f"untuk '{self.file_config}'."
            )

        except KonfigurasiError as e:

            print(f"-> GAGAL membuat PengelolaSetting: {e}")

            # Menggunakan konfigurasi kosong/default
            self._config = {}

            print(" -> Menggunakan konfigurasi default/kosong.")

    def _load_config(self):
        """
        Method internal untuk load konfigurasi.
        Bisa raise KonfigurasiError.
        """

        print(f"\n [{self.file_config}] Mencoba load konfigurasi...")

        try:
            # =====================================
            # Potensi FileNotFoundError
            # =====================================
            print(f" Membuka file '{self.file_config}'...")

            # Simulasi hanya config.txt yang valid
            if self.file_config != "config.txt":

                raise FileNotFoundError(
                    f"File '{self.file_config}' tidak ada."
                )

            # =====================================
            # Membaca isi konfigurasi
            # =====================================
            print(
                f" Membaca dan validasi isi "
                f"'{self.file_config}'..."
            )

            # =====================================
            # Simulasi Data
            # =====================================

            # DATA BENAR
            data_raw = {
                "host": "192.168.1.1",
                "port": "8080"
            }

            # DATA SALAH TIPE
            # data_raw = {
            #     "host": "192.168.1.1",
            #     "port": "xyz"
            # }

            # DATA KEY HILANG
            # data_raw = {
            #     "host": "192.168.1.1"
            # }

            # =====================================
            # Validasi Key Port
            # =====================================
            port_str = data_raw["port"]

            # Validasi tipe port
            port_int = int(port_str)

            # Simpan konfigurasi
            self._config["port"] = port_int
            self._config["host"] = data_raw.get(
                "host",
                "localhost"
            )

            print(
                " -> Konfigurasi berhasil "
                "di-load dan divalidasi."
            )

        # =====================================
        # Penanganan Error
        # =====================================

        except FileNotFoundError as e_fnf:

            print(" ERROR Bawaan: File tidak ditemukan.")

            raise KonfigurasiError(
                f"File konfigurasi penting "
                f"'{self.file_config}' hilang.",
                penyebab_asli=e_fnf
            ) from e_fnf

        except KeyError as e_key:

            print(" ERROR Bawaan: Kunci konfigurasi hilang.")

            raise KonfigurasiError(
                f"Kunci '{e_key}' wajib ada "
                f"dalam file konfigurasi.",
                penyebab_asli=e_key
            ) from e_key

        except ValueError as e_val:

            print(" ERROR Bawaan: Nilai konfigurasi tidak valid.")

            raise KonfigurasiError(
                f"Nilai untuk 'port' ('{port_str}') "
                f"harus integer.",
                penyebab_asli=e_val
            ) from e_val

        except Exception as e_lain:

            print(
                f" ERROR Bawaan Tak Terduga: "
                f"{type(e_lain).__name__}"
            )

            raise KonfigurasiError(
                "Terjadi error tak terduga "
                "saat memuat konfigurasi.",
                penyebab_asli=e_lain
            ) from e_lain

    def get_setting(self, key, default=None):
        """
        Mendapatkan nilai setting.
        """
        return self._config.get(key, default)


# =========================================
# Kode Utama
# =========================================
if __name__ == "__main__":

    # =====================================
    # Skenario 1: File Salah
    # =====================================
    print("\n--- Skenario 1: File Salah ---")

    settings1 = PengelolaSetting("setting.cfg")

    print(
        f" Port dari settings1: "
        f"{settings1.get_setting('port', 'Belum terload')}"
    )

    # =====================================
    # Skenario 2: Data Port Salah
    # =====================================
    print("\n--- Skenario 2: Data Port Salah ---")

    settings2 = PengelolaSetting("config.txt")

    print(
        f" Port dari settings2: "
        f"{settings2.get_setting('port', 'Belum terload')}"
    )

    # =====================================
    # Skenario 3: Sukses
    # =====================================
    print("\n--- Skenario 3: Sukses ---")

    settings3 = PengelolaSetting("config.txt")

    print(
        f" Port dari settings3: "
        f"{settings3.get_setting('port')}"
    )

    print(
        f" Host dari settings3: "
        f"{settings3.get_setting('host')}"
    )
```

**PROGRAM INVENTARIS BARANG DENGAN EXCEPTION HANDLING**

```python
# =========================================
# PROGRAM INVENTARIS BARANG
# DENGAN EXCEPTION HANDLING
# =========================================


# =========================================
# Kelas InventarisBarang
# =========================================
class InventarisBarang:

    def __init__(self):
        """
        Konstruktor untuk membuat inventaris kosong.
        """
        self._items = {}

        print("Inventaris Barang berhasil dibuat.")

    # =====================================
    # Method Tambah Barang
    # =====================================
    def tambah_barang(self, kode, jumlah):

        print(
            f"\n[MENAMBAH] Kode: {kode}, "
            f"Jumlah: {jumlah}"
        )

        try:
            # Validasi jumlah
            jumlah = int(jumlah)

            if jumlah <= 0:
                raise ValueError(
                    "Jumlah harus integer positif."
                )

            # Cek apakah kode sudah ada
            if kode in self._items:

                raise KeyError(
                    f"Error: Kode barang "
                    f"'{kode}' sudah ada."
                )

            # Tambahkan barang
            self._items[kode] = jumlah

            print(
                f" -> Barang '{kode}' berhasil ditambahkan."
            )

        except ValueError as e:
            print(f" -> ERROR ValueError: {e}")

    # =====================================
    # Method Ambil Barang
    # =====================================
    def ambil_barang(self, kode, jumlah):

        print(
            f"\n[MENGAMBIL] Kode: {kode}, "
            f"Jumlah: {jumlah}"
        )

        try:
            # Validasi jumlah
            jumlah = int(jumlah)

            if jumlah <= 0:
                raise ValueError(
                    "Jumlah harus integer positif."
                )

            # Cek kode barang
            if kode not in self._items:

                raise KeyError(
                    f"Error: Barang '{kode}' "
                    f"tidak ditemukan."
                )

            stok_sekarang = self._items[kode]

            # Cek stok cukup atau tidak
            if jumlah > stok_sekarang:

                raise ValueError(
                    f"Error: Stok barang '{kode}' "
                    f"tidak mencukupi "
                    f"(tersedia: {stok_sekarang}, "
                    f"diminta: {jumlah})."
                )

            # Kurangi stok
            self._items[kode] -= jumlah

            print(
                f" -> Pengambilan barang '{kode}' berhasil."
            )

            print(
                f" -> Sisa stok: "
                f"{self._items[kode]}"
            )

        except ValueError as e:
            print(f" -> ERROR ValueError: {e}")

        except KeyError as e:
            print(f" -> ERROR KeyError: {e}")

    # =====================================
    # Method Cek Stok
    # =====================================
    def cek_stok(self, kode):

        print(f"\n[CEK STOK] Kode: {kode}")

        try:
            stok = self._items[kode]

        except KeyError:
            print(
                f" -> ERROR: Barang '{kode}' "
                f"tidak ditemukan."
            )

        else:
            print(
                f" -> Stok barang '{kode}' "
                f"adalah {stok}"
            )

        finally:
            print(
                f" -> Pengecekan stok "
                f"'{kode}' selesai."
            )


# =========================================
# Kode Utama
# =========================================
if __name__ == "__main__":

    # Membuat object inventaris
    inventaris = InventarisBarang()

    # =====================================
    # Tambah Barang
    # =====================================
    print("\n===== TAMBAH BARANG =====")

    inventaris.tambah_barang("BRG001", 10)
    inventaris.tambah_barang("BRG002", 5)

    # Kode sudah ada
    try:
        inventaris.tambah_barang("BRG001", 8)

    except KeyError as e:
        print(f"Ditangkap di main: {e}")

    # Jumlah tidak valid
    inventaris.tambah_barang("BRG003", -2)

    # =====================================
    # Ambil Barang
    # =====================================
    print("\n===== AMBIL BARANG =====")

    # Pengambilan sukses
    inventaris.ambil_barang("BRG001", 3)

    # Kode tidak ada
    inventaris.ambil_barang("BRG999", 1)

    # Jumlah tidak valid
    inventaris.ambil_barang("BRG002", -5)

    # Stok tidak cukup
    inventaris.ambil_barang("BRG002", 100)

    # =====================================
    # Cek Stok
    # =====================================
    print("\n===== CEK STOK =====")

    # Barang ada
    inventaris.cek_stok("BRG001")

    # Barang tidak ada
    inventaris.cek_stok("BRG777")-
```
