**Praktikum 3: Mendesain Kelas OOP**

```python
from abc import ABC, abstractmethod

class Lokasi(ABC):

    def __init__(self, nama, latitude, longitude):
        self.nama = nama
        self.latitude = float(latitude)
        self.longitude = float(longitude)

    def get_koordinat(self):
        return (self.latitude, self.longitude)

    @abstractmethod
    def get_info_popup(self):
        pass


class TempatWisata(Lokasi):

    def __init__(self, nama, latitude, longitude, jenis, deskripsi):
        super().__init__(nama, latitude, longitude)
        self.jenis = jenis
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"{self.nama}<br>{self.jenis}<br>{self.deskripsi}"


class Kuliner(Lokasi):

    def __init__(self, nama, latitude, longitude, menu):
        super().__init__(nama, latitude, longitude)
        self.menu = menu

    def get_info_popup(self):
        return f"{self.nama}<br>Kuliner<br>{self.menu}"


class TempatIbadah(Lokasi):

    def __init__(self, nama, latitude, longitude, agama, deskripsi):
        super().__init__(nama, latitude, longitude)
        self.agama = agama
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"{self.nama}<br>{self.agama}<br>{self.deskripsi}"


print("Class berhasil dibuat")
```

**Praktikum 4: Membuat Objek dari Data**

```python
def buat_objek_lokasi_dari_df(df):

    daftar_objek = []

    for _, row in df.iterrows():

        nama = row["Nama"]
        lat = row["Latitude"]
        lon = row["Longitude"]
        tipe = row["Tipe"]
        deskripsi = row["Deskripsi"]

        if "Wisata" in tipe or tipe == "Landmark":
            objek = TempatWisata(
                nama,
                lat,
                lon,
                tipe,
                deskripsi
            )

        elif tipe == "Kuliner":
            objek = Kuliner(
                nama,
                lat,
                lon,
                deskripsi
            )

        elif "Ibadah" in tipe:
            objek = TempatIbadah(
                nama,
                lat,
                lon,
                "Umum",
                deskripsi
            )

        else:
            continue

        daftar_objek.append(objek)

    return daftar_objek


list_lokasi = buat_objek_lokasi_dari_df(df_lokasi)

for obj in list_lokasi:
    print(type(obj).__name__, "-", obj.nama)
```

```python
import folium

def buat_peta_lokasi_folium(list_objek,
                            file_output="peta_interaktif_semarang.html"):

    peta = folium.Map(
        location=[-6.9929,110.4200],
        zoom_start=12
    )

    for lokasi in list_objek:

        folium.Marker(
            location=lokasi.get_koordinat(),
            popup=lokasi.get_info_popup(),
            tooltip=lokasi.nama
        ).add_to(peta)

    peta.save(file_output)

    print("Peta berhasil dibuat")


buat_peta_lokasi_folium(list_lokasi)
```

```python
from abc import ABC, abstractmethod

class Lokasi(ABC):

    def __init__(self,nama,latitude,longitude):
        self.nama = nama
        self.latitude = float(latitude)
        self.longitude = float(longitude)

    def get_koordinat(self):
        return (self.latitude,self.longitude)

    @abstractmethod
    def get_info_popup(self):
        pass
```

```python
class TempatWisata(Lokasi):

    def __init__(self,nama,lat,lon,jenis,deskripsi):
        super().__init__(nama,lat,lon)
        self.jenis = jenis
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"<b>{self.nama}</b><br>{self.deskripsi}"


class Kuliner(Lokasi):

    def __init__(self,nama,lat,lon,menu):
        super().__init__(nama,lat,lon)
        self.menu = menu

    def get_info_popup(self):
        return f"<b>{self.nama}</b><br>{self.menu}"


class TempatIbadah(Lokasi):

    def __init__(self,nama,lat,lon,agama,deskripsi):
        super().__init__(nama,lat,lon)
        self.agama = agama
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"<b>{self.nama}</b><br>{self.deskripsi}"
```

```python
class Museum(Lokasi):

    def __init__(self,nama,lat,lon,deskripsi):
        super().__init__(nama,lat,lon)
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"<b>Museum</b><br>{self.nama}<br>{self.deskripsi}"


class Kantor(Lokasi):

    def __init__(self,nama,lat,lon,deskripsi):
        super().__init__(nama,lat,lon)
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"<b>Kantor Pemerintahan</b><br>{self.nama}<br>{self.deskripsi}"


class TamanKota(Lokasi):

    def __init__(self,nama,lat,lon,deskripsi):
        super().__init__(nama,lat,lon)
        self.deskripsi = deskripsi

    def get_info_popup(self):
        return f"<b>Taman Kota</b><br>{self.nama}<br>{self.deskripsi}"
```

```python
import pandas as pd

df = pd.read_csv("lokasi_semarang.csv")

daftar_lokasi = []

for _, row in df.iterrows():

    tipe = row["Tipe"]

    if "Wisata" in tipe or tipe == "Landmark":

        obj = TempatWisata(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            tipe,
            row["Deskripsi"]
        )

    elif tipe == "Kuliner":

        obj = Kuliner(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            row["Deskripsi"]
        )

    elif "Ibadah" in tipe:

        obj = TempatIbadah(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            "Umum",
            row["Deskripsi"]
        )

    elif tipe == "Museum":

        obj = Museum(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            row["Deskripsi"]
        )

    elif tipe == "Kantor Pemerintahan":

        obj = Kantor(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            row["Deskripsi"]
        )

    elif tipe == "Taman Kota":

        obj = TamanKota(
            row["Nama"],
            row["Latitude"],
            row["Longitude"],
            row["Deskripsi"]
        )

    daftar_lokasi.append(obj)

print("Total objek:", len(daftar_lokasi))
```

```python
try:

    with open("config_peta.txt","r") as f:

        data = f.readlines()

        lat = float(data[0].strip())
        lon = float(data[1].strip())
        zoom = int(data[2].strip())

except:

    lat = -6.9929
    lon = 110.4200
    zoom = 13

print(lat, lon, zoom)
```

```python
import folium

peta = folium.Map(
    location=[lat,lon],
    zoom_start=zoom
)

for lok in daftar_lokasi:

    warna = "blue"
    ikon = "info-sign"

    if isinstance(lok, TempatWisata):
        warna = "blue"
        ikon = "camera"

    elif isinstance(lok, Kuliner):
        warna = "red"
        ikon = "cutlery"

    elif isinstance(lok, TempatIbadah):
        warna = "green"
        ikon = "home"

    elif isinstance(lok, Museum):
        warna = "purple"
        ikon = "book"

    elif isinstance(lok, Kantor):
        warna = "gray"
        ikon = "briefcase"

    elif isinstance(lok, TamanKota):
        warna = "darkgreen"
        ikon = "leaf"

    folium.Marker(
        location=lok.get_koordinat(),
        popup=lok.get_info_popup(),
        tooltip=lok.nama,
        icon=folium.Icon(
            color=warna,
            icon=ikon
        )
    ).add_to(peta)

peta.save("peta_interaktif_semarang.html")

print("Peta berhasil dibuat")
```
