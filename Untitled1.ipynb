```python
from abc import ABC, abstractmethod

class KendaraanAbstrak(ABC):

    def __init__(self, merk):
        self.merk = merk
        print(f"Inisialisasi KendaraanAbstrak dengan merk: {self.merk}")

    def info_merk(self):
        print(f"Merk kendaraan ini adalah {self.merk}.")

    @abstractmethod
    def start_mesin(self):
        print(f"Mesin mobil {self.merk} dinyalakan.")

    @abstractmethod
    def stop_mesin(self):
        print(f"Mesin mobil {self.merk} dimatikan.")
```

```python
if __name__ == "__main__":
    print("Definisi Kelas Abstrak 'KendaraanAbstrak' selesai.")

    class Mobil(KendaraanAbstrak):
        def start_mesin(self):
            print(f"Mesin mobil {self.merk} dinyalakan.")

        def stop_mesin(self):
            print(f"Mesin mobil {self.merk} dimatikan.")

    print("\nContoh definisi kelas anak 'Mobil' selesai.")

    mobil_contoh = Mobil("Toyota")

    mobil_contoh.start_mesin()

    mobil_contoh.info_merk()

    mobil_contoh.stop_mesin()
```

Praktikum 02: Mencoba Instansiasi Kelas Abstrak

```python
from abc import ABC, abstractmethod

class MediaAbstrak(ABC):

    def __init__(self, judul):
        self.judul = judul
        print(f"Inisialisasi MediaAbstrak dengan judul: {self.judul}")

    @abstractmethod
    def play(self):
        """Metode abstrak untuk memulai pemutaran."""
        pass

    @abstractmethod
    def stop(self):
        """Metode abstrak untuk menghentikan pemutaran."""
        pass


if __name__ == "__main__":
    print("Mencoba membuat objek dari kelas abstrak MediaAbstrak...")

    try:
        media = MediaAbstrak("Konten Abstrak")

        print("Objek berhasil dibuat (SEHARUSNYA TIDAK TERJADI).")
        media.play()

    except TypeError as e:
        print(f"\nGAGAL membuat objek!")
        print(f"Error yang muncul (sesuai harapan): {e}")
        print("\nIni membuktikan bahwa kelas abstrak tidak bisa diinstansiasi")
        print("jika masih memiliki metode abstrak yang belum diimplementasikan.")
```

**Praktikum 03: Membuat Subclass Konkret - Alat Pembayaran**

```python
from abc import ABC, abstractmethod
import locale
import random


# Setting locale Indonesia
try:
    locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
except locale.Error:
    print("Locale id_ID.UTF-8 tidak tersedia, menggunakan format manual.")


# Format Rupiah aman
def format_rupiah(angka):
    return f"Rp {angka:,.0f}".replace(",", ".")


# Kelas Abstrak
class AlatPembayaranAbstrak(ABC):

    def __init__(self, nama_metode):
        self.nama_metode = nama_metode
        print(f"Inisialisasi alat pembayaran: {self.nama_metode}")

    def info(self):
        print(f"Metode Pembayaran: {self.nama_metode}")

    @abstractmethod
    def proses_pembayaran(self, jumlah):
        """
        Metode abstrak untuk memproses pembayaran.
        Harus diimplementasikan subclass.
        """
        pass


# ==================================================
# Subclass 1 : Kartu Kredit
# ==================================================
class KartuKredit(AlatPembayaranAbstrak):

    def __init__(self, nomor_kartu, nama_pemilik):
        super().__init__("Kartu Kredit")

        # Simpan 4 digit terakhir
        self.nomor_kartu = nomor_kartu[-4:]
        self.nama_pemilik = nama_pemilik

        print(
            f"-> Kartu Kredit ************{self.nomor_kartu} "
            f"({self.nama_pemilik}) siap."
        )

    # Implementasi abstract method
    def proses_pembayaran(self, jumlah):

        print(
            f"Memproses pembayaran "
            f"{format_rupiah(jumlah)} "
            f"via Kartu Kredit ************{self.nomor_kartu}..."
        )

        # Simulasi berhasil/gagal
        berhasil = random.choice([True, False])

        if berhasil:
            print("Pembayaran Kartu Kredit Berhasil.")
            return True

        else:
            print("Pembayaran Kartu Kredit Gagal (Limit/Error).")
            return False


# ==================================================
# Subclass 2 : Dompet Digital
# ==================================================
class DompetDigital(AlatPembayaranAbstrak):

    def __init__(self, nomor_telepon, nama_provider):
        super().__init__(f"Dompet Digital ({nama_provider})")

        self.nomor_telepon = nomor_telepon

        # Saldo awal random
        self._saldo = random.randint(50000, 500000)

        print(
            f"-> Dompet Digital {self.nomor_telepon} siap "
            f"(Saldo: {format_rupiah(self._saldo)})"
        )

    # Implementasi abstract method
    def proses_pembayaran(self, jumlah):

        print(
            f"Memproses pembayaran "
            f"{format_rupiah(jumlah)} "
            f"via Dompet Digital {self.nomor_telepon}..."
        )

        if jumlah <= self._saldo:

            self._saldo -= jumlah

            print("Pembayaran Dompet Digital Berhasil.")
            print(f"Saldo tersisa: {format_rupiah(self._saldo)}")

            return True

        else:
            print("Pembayaran Dompet Digital Gagal (Saldo tidak cukup).")
            print(f"Saldo saat ini: {format_rupiah(self._saldo)}")

            return False


# ==================================================
# Kode Utama
# ==================================================
if __name__ == "__main__":

    print("\n=== Membuat Objek Alat Pembayaran ===")

    kartu_bca = KartuKredit(
        "1234-5678-9012-3456",
        "Budi Cahyono"
    )

    gopay = DompetDigital(
        "08123456789",
        "GoPay"
    )

    print("\n=== Melakukan Pembayaran ===")

    # Pembayaran kartu kredit
    print("\n[Mencoba bayar dengan Kartu Kredit]")
    kartu_bca.info()

    status_kk = kartu_bca.proses_pembayaran(150000)

    print(
        f"Status Transaksi KK: "
        f"{'Sukses' if status_kk else 'Gagal'}"
    )

    # Pembayaran GoPay kecil
    print("\n[Mencoba bayar dengan GoPay - jumlah kecil]")
    gopay.info()

    status_gopay1 = gopay.proses_pembayaran(75000)

    print(
        f"Status Transaksi GoPay 1: "
        f"{'Sukses' if status_gopay1 else 'Gagal'}"
    )

    # Pembayaran GoPay besar
    print("\n[Mencoba bayar dengan GoPay - jumlah besar]")
    gopay.info()

    status_gopay2 = gopay.proses_pembayaran(1000000)

    print(
        f"Status Transaksi GoPay 2: "
        f"{'Sukses' if status_gopay2 else 'Gagal'}"
    )
```

**Praktikum 04: Menggunakan Kelas Abstrak untuk Polimorfisme**

```python
from abc import ABC, abstractmethod
import time


# ==================================================
# Kelas Abstrak
# ==================================================
class DokumenAbstrak(ABC):

    def __init__(self, nama_file):
        self.nama_file = nama_file
        print(f"Inisialisasi Dokumen: {self.nama_file}")

    def info_file(self):
        print(f"Nama File: {self.nama_file}")

    @abstractmethod
    def cetak(self):
        """Metode abstrak untuk mencetak dokumen."""
        pass

    @abstractmethod
    def simpan(self):
        """Metode abstrak untuk menyimpan dokumen."""
        pass


# ==================================================
# Subclass Konkret 1
# ==================================================
class DokumenTeks(DokumenAbstrak):

    def __init__(self, nama_file, isi_teks):
        super().__init__(nama_file)
        self.isi_teks = isi_teks

        print("-> Tipe: Dokumen Teks")

    def cetak(self):

        print(f"\nMencetak Dokumen Teks '{self.nama_file}':")
        print("=" * 20)

        print(self.isi_teks)

        print("=" * 20)

    def simpan(self):

        print(f"Menyimpan Dokumen Teks '{self.nama_file}' ke disk...")

        # Simulasi penyimpanan
        time.sleep(0.2)

        print("-> Berhasil disimpan.")


# ==================================================
# Subclass Konkret 2
# ==================================================
class Spreadsheet(DokumenAbstrak):

    def __init__(self, nama_file, jumlah_baris, jumlah_kolom):
        super().__init__(nama_file)

        self.baris = jumlah_baris
        self.kolom = jumlah_kolom

        print("-> Tipe: Spreadsheet")

    def cetak(self):

        print(
            f"\nMencetak Spreadsheet "
            f"'{self.nama_file}' "
            f"({self.baris} baris x {self.kolom} kolom)..."
        )

        print("(Menampilkan pratinjau spreadsheet...)")

    def simpan(self):

        print(
            f"Menyimpan Spreadsheet "
            f"'{self.nama_file}' "
            f"ke format .xlsx..."
        )

        # Simulasi penyimpanan
        time.sleep(0.3)

        print("-> Berhasil disimpan.")


# ==================================================
# Fungsi Polimorfik
# ==================================================
def proses_dokumen(daftar_dokumen):

    print("\n======= MEMPROSES SEMUA DOKUMEN =======")

    for doc in daftar_dokumen:

        print(
            f"\n--- Memproses "
            f"{type(doc).__name__}: "
            f"{doc.nama_file} ---"
        )

        try:
            # Polymorphism
            doc.cetak()
            doc.simpan()

        except Exception as e:
            print(f"Error saat memproses {doc.nama_file}: {e}")

    print("\n======= SELESAI MEMPROSES DOKUMEN =======")


# ==================================================
# Kode Utama
# ==================================================
if __name__ == "__main__":

    print("Membuat Dokumen...\n")

    dok1 = DokumenTeks(
        "laporan.txt",
        "Ini adalah isi laporan singkat."
    )

    dok2 = Spreadsheet(
        "data_penjualan.xlsx",
        200,
        15
    )

    dok3 = DokumenTeks(
        "catatan_rapat.txt",
        "Poin penting:\n- Bahas budget\n- Tentukan timeline"
    )

    koleksi_dokumen = [
        dok1,
        dok2,
        dok3
    ]

    # Pemrosesan polymorphism
    proses_dokumen(koleksi_dokumen)

    # Contoh akses atribut spesifik
    # print(dok1.isi_teks)
```

**Praktikum 05: Kelas Abstrak dengan Metode Konkret dan Abstrak**

```python
from abc import ABC, abstractmethod
import math


# ==================================================
# Kelas Abstrak
# ==================================================
class BangunDatarAbstrak(ABC):

    def __init__(self, nama):

        # Protected attribute
        self._nama = nama

        print(f"Inisialisasi BangunDatarAbstrak: {self._nama}")

    # ==================================================
    # Metode Konkret
    # ==================================================

    @property
    def nama(self):
        """Getter untuk nama bangun datar."""
        return self._nama

    def info_lengkap(self):
        """
        Menampilkan informasi lengkap bangun datar.
        """

        print(f"\n--- Info Lengkap: {self.nama} ---")

        # Hitung luas
        try:
            luas = self.hitung_luas()

            print(f"Luas      : {luas:.2f}")

        except NotImplementedError:
            print("Luas      : Belum diimplementasikan.")

        except Exception as e:
            print(f"Luas      : Error ({e})")

        # Hitung keliling
        try:
            keliling = self.hitung_keliling()

            print(f"Keliling  : {keliling:.2f}")

        except NotImplementedError:
            print("Keliling  : Belum diimplementasikan.")

        except Exception as e:
            print(f"Keliling  : Error ({e})")

        print("-" * (len(self.nama) + 20))

    # ==================================================
    # Metode Abstrak
    # ==================================================

    @abstractmethod
    def hitung_luas(self):
        """Menghitung luas."""
        pass

    @abstractmethod
    def hitung_keliling(self):
        """Menghitung keliling."""
        pass


# ==================================================
# Subclass Konkret
# ==================================================
class Lingkaran(BangunDatarAbstrak):

    def __init__(self, radius):

        # Memanggil constructor parent
        super().__init__("Lingkaran")

        if radius < 0:
            raise ValueError("Radius tidak boleh negatif")

        self.radius = radius

        print(f"-> Lingkaran dibuat (Radius: {self.radius})")

    # Implementasi abstract method
    def hitung_luas(self):
        return math.pi * (self.radius ** 2)

    # Implementasi abstract method
    def hitung_keliling(self):
        return 2 * math.pi * self.radius


# ==================================================
# Kode Utama
# ==================================================
if __name__ == "__main__":

    print("Membuat objek Lingkaran...\n")

    lingkaran_A = Lingkaran(10)

    # Mengakses property
    print(f"\nNama bangun (via property): {lingkaran_A.nama}")

    # Memanggil method konkret parent
    lingkaran_A.info_lengkap()

    # Contoh panggil langsung
    # luas = lingkaran_A.hitung_luas()
    # keliling = lingkaran_A.hitung_keliling()

    # print(f"Luas langsung: {luas:.2f}")
    # print(f"Keliling langsung: {keliling:.2f}")
```

**Praktikum 06: Properti Abstrak - Elemen Grafis**

```python
from abc import ABC, abstractmethod


# ==================================================
# Kelas Abstrak
# ==================================================
class ElemenGrafis(ABC):

    def __init__(self, id_elemen, warna):
        print(f"[ElemenGrafis __init__] Membuat elemen dengan ID: {id_elemen}")

        self._id = id_elemen
        self.warna = warna

    # ==================================================
    # Properti Konkret (Getter)
    # ==================================================
    @property
    def id_elemen(self):
        print(f"[ElemenGrafis id_elemen getter] Mengakses ID: {self._id}")

        return self._id

    # ==================================================
    # Properti Abstrak : posisi
    # ==================================================

    # Getter abstrak
    @property
    @abstractmethod
    def posisi(self):
        """
        Subclass harus mendefinisikan cara mendapatkan posisi.
        """
        print("[ElemenGrafis posisi getter abstract] Harus dioverride.")
        ...

    # Setter abstrak
    @posisi.setter
    @abstractmethod
    def posisi(self, koordinat_baru):
        """
        Subclass harus mendefinisikan cara mengubah posisi.
        """
        print("[ElemenGrafis posisi setter abstract] Harus dioverride.")
        ...

    # ==================================================
    # Metode Abstrak
    # ==================================================
    @abstractmethod
    def gambar(self):
        """
        Subclass harus mendefinisikan cara menggambar elemen.
        """
        print("[ElemenGrafis gambar abstract] Harus dioverride.")
        pass

    # ==================================================
    # Metode Konkret
    # ==================================================
    def info_warna(self):
        print(
            f"[ElemenGrafis info_warna] "
            f"Warna elemen {self.id_elemen}: {self.warna}"
        )


# ==================================================
# Subclass Konkret
# ==================================================
class Kotak(ElemenGrafis):

    def __init__(self, id_elemen, warna, x=0, y=0, lebar=10, tinggi=10):

        # Panggil constructor parent
        super().__init__(id_elemen, warna)

        print(f"[Kotak __init__] Menginisialisasi Kotak ID: {self.id_elemen}")

        # Posisi internal
        self._x = x
        self._y = y

        self.lebar = lebar
        self.tinggi = tinggi

    # ==================================================
    # Implementasi Getter Properti Abstrak
    # ==================================================
    @property
    def posisi(self):

        print(
            f"[Kotak posisi getter] "
            f"Mengembalikan posisi Kotak {self.id_elemen}: "
            f"({self._x}, {self._y})"
        )

        return (self._x, self._y)

    # ==================================================
    # Implementasi Setter Properti Abstrak
    # ==================================================
    @posisi.setter
    def posisi(self, koordinat_baru):

        print(
            f"[Kotak posisi setter] "
            f"Mencoba set posisi Kotak {self.id_elemen} "
            f"ke {koordinat_baru}"
        )

        if isinstance(koordinat_baru, tuple) and len(koordinat_baru) == 2:

            self._x = koordinat_baru[0]
            self._y = koordinat_baru[1]

            print(
                f"-> Posisi Kotak {self.id_elemen} "
                f"berhasil diubah ke ({self._x}, {self._y})"
            )

        else:
            print("-> Gagal: Posisi harus berupa tuple (x, y).")

    # ==================================================
    # Implementasi Metode Abstrak
    # ==================================================
    def gambar(self):

        print(
            f"[Kotak gambar] Menggambar Kotak '{self.id_elemen}' "
            f"warna {self.warna} di ({self._x}, {self._y}) "
            f"dengan ukuran {self.lebar} x {self.tinggi}"
        )


# ==================================================
# Kode Utama
# ==================================================
if __name__ == "__main__":

    print("Membuat objek Kotak...\n")

    kotak1 = Kotak(
        "KotakA",
        "Merah",
        x=5,
        y=10,
        lebar=20
    )

    print("-" * 30)

    # Mengakses properti konkret
    print("Mengakses ID...")
    id_ktk = kotak1.id_elemen

    print("-" * 30)

    # Mengakses properti abstrak
    print("Mengakses Posisi Awal...")
    pos_awal = kotak1.posisi

    print("-" * 30)

    # Mengubah posisi
    print("Mengubah Posisi...")
    kotak1.posisi = (50, 60)

    print("-" * 30)

    # Mengakses posisi baru
    print("Mengakses Posisi Baru...")
    pos_baru = kotak1.posisi

    print("-" * 30)

    # Contoh input salah
    print("Mencoba Set Posisi Salah...")
    kotak1.posisi = [100, 200]

    print("-" * 30)

    # Menggambar kotak
    print("Menggambar Kotak...")
    kotak1.gambar()

    print("-" * 30)

    # Method konkret parent
    print("Info Warna...")
    kotak1.info_warna()

    print("-" * 30)
```

**Program Hierarki Kelas Senjata Menggunakan Konsep Abstract Base Class**

```python
from abc import ABC, abstractmethod


# ==================================================
# Kelas Abstrak
# ==================================================
class Senjata(ABC):

    def __init__(self, nama):
        self.nama = nama

    # Method konkret
    def info_nama(self):
        print(f"Nama Senjata: {self.nama}")

    # Method abstrak
    @abstractmethod
    def serang(self):
        pass

    # Properti abstrak
    @property
    @abstractmethod
    def ammo(self):
        pass


# ==================================================
# Kelas Konkret 1 : Pedang
# ==================================================
class Pedang(Senjata):

    def __init__(self, nama, panjang_bilah):
        super().__init__(nama)

        self.panjang_bilah = panjang_bilah
        self._daya_tahan = 100

    # Implementasi method abstrak
    def serang(self):
        print(
            f"Pedang {self.nama} menebas "
            f"dengan bilah sepanjang "
            f"{self.panjang_bilah} cm!"
        )

    # Implementasi property abstrak
    @property
    def ammo(self):
        return f"Daya tahan: {self._daya_tahan}%"


# ==================================================
# Kelas Konkret 2 : Panah
# ==================================================
class Panah(Senjata):

    def __init__(self, nama, jumlah_anak_panah):
        super().__init__(nama)

        self._jumlah_anak_panah = jumlah_anak_panah

    # Implementasi method abstrak
    def serang(self):

        if self._jumlah_anak_panah > 0:

            self._jumlah_anak_panah -= 1

            print(
                f"Panah {self.nama} melesat! "
                f"Sisa anak panah: "
                f"{self._jumlah_anak_panah}"
            )

        else:
            print(f"Amunisi Panah {self.nama} habis!")

    # Implementasi property abstrak
    @property
    def ammo(self):
        return self._jumlah_anak_panah

    # Setter property
    @ammo.setter
    def ammo(self, jumlah_baru):

        if jumlah_baru >= 0:
            self._jumlah_anak_panah = jumlah_baru

            print(
                f"Panah {self.nama} diisi ulang. "
                f"Jumlah ammo sekarang: "
                f"{self._jumlah_anak_panah}"
            )

        else:
            print("Jumlah ammo tidak valid!")


# ==================================================
# Kode Utama
# ==================================================
if __name__ == "__main__":

    print("=== MEMBUAT SENJATA ===")

    pedang1 = Pedang("Excalibur", 120)
    panah1 = Panah("Dragon Bow", 3)

    print("\n=== SERANGAN PEDANG ===")
    pedang1.info_nama()
    pedang1.serang()

    print(f"{pedang1.ammo}")

    print("\n=== SERANGAN PANAH ===")

    panah1.info_nama()

    # Menyerang beberapa kali
    panah1.serang()
    panah1.serang()
    panah1.serang()
    panah1.serang()

    print(f"Sisa Ammo: {panah1.ammo}")

    print("\n=== ISI ULANG PANAH ===")

    panah1.ammo = 5

    print(f"Ammo Sekarang: {panah1.ammo}")

    # Optional
    print("\n=== UJI ABSTRACT CLASS ===")

    try:
        senjata_umum = Senjata("Senjata Misterius")

    except TypeError as e:
        print("Gagal membuat objek dari abstract class!")
        print(f"Error: {e}")
```
